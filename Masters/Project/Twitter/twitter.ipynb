```python
# Authentication with Twitter API
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
```

```python
# Creating an API object to access Twitter
api = tweepy.API(auth)
```

```python
# Define search query and number of tweets to extract
search_words = "absa"
date_since = "2022-01-01"
date_until = "2022-12-31"
num_of_tweets = 1500 
```

```python
# Collect tweets using Cursor object
tweets = tweepy.Cursor(api.search_tweets,
                       q=search_words,
                       lang="en",
                       since_id=date_since,
                       until=date_until).items(num_of_tweets)
```

```python
# Creating a list of dictionaries containing tweet data
tweets_list = []
for tweet in tweets:
    tweets_list.append({
        "tweet_id": tweet.id_str,
        "username": tweet.user.screen_name,
        "text": tweet.text,
        "created_at": tweet.created_at,
        "retweet_count": tweet.retweet_count,
        "favorite_count": tweet.favorite_count,
        "source": tweet.source
    })
```

```python
# Converting list of dictionaries to Pandas DataFrame
absa_absa_tweets_df = pd.DataFrame(tweets_list)


#Save raw data file in csv
absa_absa_tweets_df.to_csv('C:/Miehleketo/msc-miehleketo/data/raw/Absa_twitter.csv', index=False)

absa_absa_tweets_df.to_csv('C:/Miehleketo/msc-miehleketo/src/data/Absa_twitter.csv', index=False)
```
